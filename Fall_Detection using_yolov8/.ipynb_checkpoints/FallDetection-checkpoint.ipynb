```python
# Use a lock to control thread creation
thread_lock = threading.Lock()
alert_sent = {}

video = cv2.VideoCapture(0)
print(video.isOpened())

def process_frame(frame):
    results = model(frame)[0]
    boxes = results.boxes  # Boxes object
    confs = boxes.conf.cpu().numpy()  # Confidence scores
    class_ids = boxes.cls.cpu().numpy()  # Class IDs

    for box, conf, cls_id in zip(boxes.xyxy.cpu().numpy(), confs, class_ids):
        if conf >= 0.78:
            x1, y1, x2, y2 = map(int, box)
            label = results.names[int(cls_id)]

            # Draw the bounding box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            # Put the label above the bounding box
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

            # Check if the label is "FallDetected" and not already alerted
            if label == "FallDetected" and label not in alert_sent:
                # Acquire lock before creating thread
                with thread_lock:
                    # Send email in a separate thread
                    threading.Thread(target=send_email_in_thread, args=(label,)).start()
                    alert_sent[label] = True

    # Display the frame with detections
    cv2.imshow('objects', frame)

# Main loop
while video.isOpened():
    ret, frame = video.read()
    if not ret:
        break

    # Process frame in a separate function
    process_frame(frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
def list_available_cameras(max_cameras=10):
    available_cameras = []
    for camera_id in range(max_cameras):
        cap = cv2.VideoCapture(camera_id)
        if cap.isOpened():
            available_cameras.append(camera_id)
            cap.release()
    return available_cameras

# List all available cameras
cameras = list_available_cameras()
print(f"Available cameras: {cameras}")
```
